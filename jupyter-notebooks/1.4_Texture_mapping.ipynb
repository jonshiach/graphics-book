# Texture mapping

```python
import numpy as np
import matplotlib.pyplot as plt     
from myst_nb import glue
plt.rcParams['text.usetex'] = True
import matplotlib.image as mpimg


            
def texturemapper(R, T, xv, yv):
    
    Nv, Nu, _ = T.shape
    uv, vv = [0, Nu, Nu, 0], [Nv, Nv, 0, 0]
    
    # Generate edge table
    n = len(xv)
    ET = []
    j = n - 1
    for i in range(n):
        if yv[i] == yv[j]:
            j = i
            continue
        elif yv[i] < yv[j]:
            imin, imax = i, j
        else:
            imin, imax = j, i
        xmin, xmax, ymin, ymax = xv[imin], xv[imax], yv[imin], yv[imax]
        m = (xmin - xmax) / (ymin - ymax)
        umin, umax, vmin, vmax = uv[imin], uv[imax], vv[imin], vv[imax]
        du = umin - umax
        dv = vmin - vmax
        ET.append([xmin, ymin, ymax, m, umin, vmin, du, dv])
        j = i
        
    # Loop through scanlines
    y = min(yv)
    AET = []
    while True:
        
        # Move edges from ET to AET whose ymin = y
        i = 0
        while i < len(ET):
            if ET[i][1] == y:
                AET.append(ET.pop(i))
            else:
                i += 1
            
        # Remove edges from AET whose ymax = y
        for edge in AET.copy():
            if edge[2] == y:
                AET.remove(edge)
        
        # Check for completion
        if len(ET) == 0 and len(AET) == 0:
            break
        
        # Sort AET by x vale
        AET.sort()
        
        # Fill scanline
        i = 0
        while i < len(AET) - 1:
            xL = int(AET[i][0]) + 1
            xR = int(AET[i+1][0]) + 1
            u = int(AET[i][4])
            v = int(AET[i][5])
            j = 0
            while u < Nu and v < Nv:
                
            
            R[y,xL:xR,:] = [255, 0, 0]
            i += 2
            
        # Increment y
        y += 1
        
        # Update x values in AET
        for edge in AET:
            edge[0] += edge[-1]
        
    return R 
    

blue = [210, 210, 255]
darkblue = [0, 0, 255]
grey = [128, 128, 128]

Nx, Ny = 20, 20
img = 255 * np.ones((Nx, Ny, 3)).astype(int)

fig, ax = plt.subplots()

imgplot = plt.imshow(img)
for i in range(Nx):
    for j in range(Ny):
        ax.add_patch(plt.Rectangle((i-0.5,j-0.5), 1, 1, ec="k", fill=False))
       
x = [2, 16, 14, 4]
y = [16, 16, 6, 2]


texture = mpimg.imread("../images/cavendish.jpg")

img = texturemapper(img, texture, x, y)

plt.imshow(img)
plt.fill(x, y, fill=False,lw=2)
   
plt.xticks([])
plt.yticks([])
glue("scanline-raster", fig, display=False)
```
